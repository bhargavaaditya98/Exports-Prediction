```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,20,3)
y=np.linspace(1,10,7)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'aqua')
plt.xlabel('xvalues')
plt.ylabel('yvalues')
plt.title('x and y')
```

```python
x=[1,2,3,4,5,8]
y=[4,2,5,8,6,7]
plt.scatter(x,y,color='aqua')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x and y')
```

```python
x1=[1,2,3,4,5,8]
y1=[4,2,5,8,6,7]
plt.plot(x,y,color='aqua')
plt.plot(x1,y1,color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x and y')
```

```python
x=np.arange(1,20,3)
y=np.linspace(1,10,7)

plt.bar(x,y,color='aqua',label='plot1')
plt.xlabel('xvalues')
plt.ylabel('yvalues')
plt.title('x and y')
x=[1,2,3,4,5,8]
y=[4,2,5,8,6,7]
plt.bar(x,y,color='blue',label='plot2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x and y')
plt.legend()
```

```python
list=[60,70,75,80,60]
labels=["maths","physics","chem","bio","eng"]
plt.pie(list,labels=labels)
plt.legend()
plt.show()
```

```python
x=np.linspace(1,10,8)
y=x**2
plt.subplot(2,2,1)
plt.plot(x,y,'red')
plt.subplot(2,2,2)
plt.plot(x1,y1)
plt.subplot(2,2,3)
plt.plot(x,y,'red')
plt.subplot(2,2,4)
plt.plot(x1,y1)
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.2,0.3])
axes.plot(x,y,'aqua')
axes.set_xlabel("hh")

axes.set_ylabel("zxc")
axes.set_title("asdfg")
axes2.plot(x,y,'aqua')
axes2.set_xlabel("hh")
axes2.set_ylabel("zxc")
axes2.set_title("asdfg")
```
