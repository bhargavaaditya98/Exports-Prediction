```python
import pandas as pd  #importing libraries pandas,numpy,matplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("exports.csv") #reading the dataset
```

```python
data.head(10) #verify the dataset
```

```python
data.isnull().any() #checks for number of null = NaN
```

```python
drop = ['fruits fresh','fruits proc','veggies fresh','veggies proc'] #array headings of not required attributes
```

```python
data.drop(drop,inplace = True,axis = 1) #drop headings with null values 
```

```python
data #new data after removal of all unwanted data
```

```python
x = data.iloc[:,4:13]#declared the independent variables
```

```python
x
```

```python
y = data.iloc[:,3]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size = 0.25,random_state = 0) #split the data into data train and data test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()#created our regressor
regressor.fit(x_train, y_train)
```

```python
ypred = regressor.predict(x_test)#predict the new values from the old x test
```

```python
ypred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, ypred) #evaluates the accuracy of the model
```
